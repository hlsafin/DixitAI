```python
import csv
import os
import json
```

```python
ANNOTATIONS_URL = './annotations/'
annotation_dict = dict()
```

```python
for f in os.listdir(ANNOTATIONS_URL):
    with open(ANNOTATIONS_URL+f, newline='') as csvfile:
        image_num = int(f.split('.')[0])
        annotation_dict[image_num] = dict()
        
        annotation_reader = csv.reader(csvfile, delimiter=',', quotechar='\'')
        
        for row in annotation_reader:
            if(row[0] == 'image_num'):
                annotation_dict[image_num][row[0]] = int(row[1])
            else:
                annotation_dict[image_num][row[0]] = list(map(str.strip, row[1:]))
```

```python
annotation_dict
```

```python
with open('all_annotations.json', 'w', encoding='utf-8') as f:
    json.dump(annotation_dict, f, ensure_ascii=False, indent=4)
```
