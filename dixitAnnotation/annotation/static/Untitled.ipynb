```python
import csv
import os
import pandas as pd
```

```python
ANNOTATIONS_URL = './annotations/'
annotation_count = dict()
n_files = 0
```

```python
for f in os.listdir(ANNOTATIONS_URL):
    with open(ANNOTATIONS_URL+f, newline='') as csvfile:
        n_files += 1
        image_num = int(f.split('.')[0])
        annotation_dict[image_num] = dict()
        
        annotation_reader = csv.reader(csvfile, delimiter=',', quotechar='\'')
        
        for row in annotation_reader:
            if(row[0] != 'image_num'):
                for ann in list(map(str.strip, row[1:])):
                    for word in ann.split():
                        if(word == ''):
                            continue
                        if(word.lower() in annotation_count):
                            annotation_count[word.lower()] += 1
                        else:
                            annotation_count[word.lower()] = 1
```

```python
for key in list(sorted(annotation_count, key=lambda x:annotation_count[x], reverse=True)):
    with open('frequency_ordered_human_annotation.csv', 'w', newline='') as csvfile:
        csvwriter = csv.writer(csvfile, delimiter='\t',
                                quotechar='\'', quoting=csv.QUOTE_MINIMAL)
        csvwriter.writerow([key, annotation_count[key]/n_files])
    print(key, annotation_count[key]/n_files)
```
